```python
import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
xOne = np.random.rand(5,1) *10 -5
yOne = 2 * xOne + 5
yOne = yOne + np.random.normal(scale=1, size=yOne.shape)
plt.plot(xOne, yOne, 'bo')
```

```python
polyFeatures = PolynomialFeatures(degree=1, include_bias=False)
xOnePoly = polyFeatures.fit_transform(xOne)
reg1_1 = LinearRegression()
reg1_1.fit(xOnePoly, yOne)
```

```python
xOneVals = xOne
xOneValsPoly = polyFeatures.transform(xOneVals)
yOneVals = reg1_1.predict(xOneValsPoly)
plt.suptitle('DATA 1 : Degree 1')
plt.ylabel('Output')
plt.xlabel('Input')
plt.ylim((-20,20))
plt.xlim((-5,5))
plt.scatter(xOne, yOne)
plt.plot(xOneVals, yOneVals, color="lightGreen")
plt.show()
```

```python
print(f" intercept : {reg1_1.intercept_} \n coefficient : {reg1_1.coef_} \n MSE : {metrics.mean_squared_error(yOne, yOneVals)} \n R2 Score : {r2_score(yOne, yOneVals)}")
```

```python
polyFeatures_4 = PolynomialFeatures(degree=4, include_bias=False)
xOnePoly_4 = polyFeatures_4.fit_transform(xOne)
reg1_1_4 = LinearRegression()
reg1_1_4.fit(xOnePoly_4, yOne)
```

```python
xOneVals_4_show = np.linspace(-5,5,500).reshape(-1,1)
xOneValsPoly_4_show = polyFeatures_4.transform(xOneVals_4_show)
yOneVals_4_show = reg1_1_4.predict(xOneValsPoly_4_show)
xOneVals_4 = xOne
xOneValsPoly_4 = polyFeatures_4.transform(xOneVals_4)
yOneVals_4 = reg1_1_4.predict(xOneValsPoly_4)
plt.suptitle('DATA 1 : Degree 4')
plt.ylabel('Output')
plt.xlabel('Input')
plt.ylim((-20,20))
plt.xlim((-5,5))
plt.scatter(xOne, yOne)
plt.plot(xOneVals_4_show, yOneVals_4_show, color="navy")
plt.show()
```

```python
print(f" intercept : {reg1_1_4.intercept_} \n coefficient : {reg1_1_4.coef_} \n MSE : {metrics.mean_squared_error(yOne, yOneVals_4)} \n R2 Score : {r2_score(yOne, yOneVals_4)}")
```
